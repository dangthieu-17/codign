W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a, b = map(int, input().split())

if a <= 0 or b <= 0:
    print("False")
else:
    c = min(a, b)
    for i in range(1, c + 1):
        if a % i == 0 and b % i == 0:
            print(i, end=" ")
```

W4A22 Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số chẵn, bao nhiêu chữ số lẻ

```python
n = int(input())
chan = 0
le = 0
while n > 0:
    a = n % 10
    if a % 2 == 0:
        chan += 1
    else:
        le += 1
    n = n // 10
print(f"{chan} {le}")
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 

```python
n = int(input())
max = n
min = n
while n != -1:
    if n > max:
        max = n
    else:
        min = n  
    n = int(input()) 
print(f"{max} {min}")
```

W4A27
Nhập vào một chuỗi, hãy đếm xem trong chuỗi có bao nhiêu từ (quy định là chuỗi không có
ký tự đặc biệt, không số, không có dấu câu, chỉ có ký tự chữ và k

```python
s = input("Nhập chuỗi: ")

count = 0
in_word = False

for char in s:
    if char != " ":
        if not in_word:
            count += 1
            in_word = True 
    else:
        in_word = False    

print(f"Số từ trong chuỗi là: {count}")
```
