Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu
vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2
chữ số phần thập phân. 

```python
a = float(input("chieu rong a: "))
b = float(input("chieu rong b: "))
pi = 3.14
r = a / 2 
S_hcn = a * b
S_c = pi * r * r

S = S_hcn - S_c 
print(round(S, 2))
```

Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case 

```python
n = int(input())
match n:
    case 0:
        print("so khong")
    case 1:
        print("so mot")
    case 2:
        print("so hai")
    case 3:
        print("so ba")
    case 4:
        print("so bon")
    case _:
        print("ERor")
```
