Ngược lại, kiểm tra xem nó là tam giác gì: Đều, Cân, Vuông, hay Tam giác thường. Tính diện tích, chu vi của tam giác và in ra mànhình.

```python
a, b, c = map(float, input().split())
if a + b > c and a + c > b and b + c > a:
    if a == b == c:
        print("đây là tam giác đều")
    elif a*a + b*b == c*c or a*a + c*c == b*b or b*b + c*c == a*a:
        print("đây là tam giác vuông")
    elif a == b or a == c or c == b:
        print("đây là tam giác cân")
    else:
        print("đây là tam giác thường")
# chu vi tam giac
P = a + b + c 
print(P)
p = P/2
S = math.sqrt(p * (p - a) * (p - b) * (p - c))
print(S)
```

Bài 5: Viết chương trình giải phương trình bậc 2 (ax2 + bx + c = 0)

```python
a, b, c = map(int,input().split())
if a == 0:
    if b == 0:
        if c == 0:
            print("Phương trình vô số nghiệm")
        else:
            print("Phương trình vô nghiệm")
    else:
        print(f"Phương trình bậc 1 có nghiệm x = {-c/b}")
else:
    # Tính Delta
    delta = b**2 - 4*a*c
    
    if delta < 0:
        print("Phương trình vô nghiệm")
    elif delta == 0:
        x = -b / (2*a)
        print(f"Phương trình có nghiệm x = {x}")
    else: 
        x1 = (-b + delta**0.5) / (2*a)
        x2 = (-b - delta**0.5) / (2*a)
        print(f"Phương trình có 2 nghiệm phân biệt:")
        print(f"x1 = {x1}")
        print(f"x2 = {x2}")
```
